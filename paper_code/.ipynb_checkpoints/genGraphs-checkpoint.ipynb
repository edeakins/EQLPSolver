```python
import pandas as pd
import matplotlib.pyplot as plt
mLibDat = pd.read_csv("/mnt/c/Users/edeak/Mittleman_Timings_Paper.csv")
mLibDf = pd.DataFrame(mLibDat)
```

```python
sTime = dict(zip(mLibDf.loc[:,'Total Time HiGHS'], mLibDf.loc[:,'Total Time']))
creducs = dict(zip(mLibDf.loc[:,'Total Time HiGHS'], mLibDf.loc[:,'Column Reduction (%)']))
rreducs = dict(zip(mLibDf.loc[:,'Total Time HiGHS'], mLibDf.loc[:,'Row Reduction (%)']))
nreducs = dict(zip(mLibDf.loc[:,'Total Time HiGHS'], mLibDf.loc[:,'Nonzero Reduction (%)']))
```

```python
plotPoints = {}
idx = 0
x = []
y = []
for key, val in sTime.items():
    if key < 5: 
        continue
    gainloss = ((key - val)/key)*100
    x.append(creducs[key])
    y.append(gainloss)
```

```python
plt.scatter(x,y,c="black")
plt.xlabel("Column Reduction (%)")
plt.ylabel("Solve Time Gain/Loss (%)")
# plt.title("Solve Time Gain/Loss Compared to Column Reduction")
plt.savefig("colRed.png")
```
